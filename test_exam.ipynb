# 1 Aufgabe: Grundlagen

```python
import numpy as np
import matplotlib.pyplot as plt
```

Parameter aus Haupt Bodendynamik 5.3.1. Beispiel: Vertikal erregtes Blockfundament

```python
k = 382e6
m = 8.5e3
zeta = 0.48
Fhat = 5.9e3
```

```python
omega0 = np.sqrt(k/m)
f0 = omega0/(2*np.pi)
c = 2*zeta*omega0*m
omega1 = omega0*np.sqrt(1-zeta**2)
print("f0 = {} Hz".format(f0))
print("c = {} Ns/m".format(c))
print("omega1 = {} RAD/s".format(omega1))
```

```python
a1 = Fhat/k
f1 = 30
a2 = 0.9*Fhat/k
f2 = 40
```

```python
T = 1/f1
t = np.linspace(0, 2*T)
a1t = a1*np.sin(2*np.pi*f1*t) 
a2t = a2*np.sin(2*np.pi*f2*t) 
plt.plot(t, a1t, 'k--', t, a2t, 'k--', t, a1t+a2t, 'k');
```

```python
eta = f1/f0
V = 1/((1-eta**2)**2+(2*zeta*eta)**2)
psi = np.arctan2(2*zeta*eta, 1-eta**2)
```

```python
psi
```

```python
w1 = 2*np.pi*f1
UP1 = V*a1
up1 = UP1*np.sin(w1*t - psi)
plt.plot(t, a1t, 'r', t, up1);
```

```python
VP1=UP1*w1
Ekinmax=(1/2)*m*VP1**2
print("Ekinmax = {} J".format(Ekinmax))
```

```python
vp1 = VP1*np.cos(w1*t - psi)
FF = k*up1
FD = c*vp1
plt.plot(t, FF, 'r', t, FD, 'g', t, FF+FD, 'b');
```
